```python
categorical_cols = data.select_dtypes(include=['object']).columns
category_counts = {col: data[col].nunique() for col in categorical_cols}
```

```python
onehot_cols = [col for col, count in category_counts.items() if count < 5]
data_encoded = pd.get_dummies(data, columns=onehot_cols, prefix=onehot_cols)
```

```python
label_cols = [col for col, count in category_counts.items() if count >= 5]
for col in label_cols:
    le = LabelEncoder()
    data_encoded[col] = le.fit_transform(data_encoded[col])
```

```python
# Combined Education and Occupation
data_encoded['education_occupation'] = data_encoded['education'].astype(str) + '_' + data_encoded['occupation'].astype(str)

# Hours Worked Per Week Category


# Print the available columns to identify the correct column name
print(data_encoded.columns)

# Assuming the column name is actually 'hours_per_week' (replace if different)
data_encoded['hours_worked_category'] = pd.cut(data_encoded['hours_per_week'], bins=[0, 30, 40, float('inf')], labels=['part-time', 'full-time', 'overtime'])
```

```python
import numpy as np # Import the numpy library and assign it the alias 'np'

# Check if 'capital-gain' is in the columns
if 'capital-gain' in data_encoded.columns:
    data_encoded['capital-gain'] = np.log1p(data_encoded['capital-gain'])
else:
    print("Warning: 'capital-gain' column not found in data_encoded.")

# Check if 'capital-loss' is in the columns
if 'capital-loss' in data_encoded.columns:
    data_encoded['capital-loss'] = np.log1p(data_encoded['capital-loss'])
else:
    print("Warning: 'capital-loss' column not found in data_encoded.")
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.ensemble import IsolationForest
import numpy as np

# List of categorical columns to encode
categorical_cols = data_encoded.select_dtypes(include=['object', 'category']).columns

# Create a OneHotEncoder instance
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')  # sparse=False for dense output

# Fit and transform the categorical features
encoded_data = encoder.fit_transform(data_encoded[categorical_cols])

# Create a new DataFrame with the encoded features
encoded_df = pd.DataFrame(encoded_data, columns=encoder.get_feature_names_out(categorical_cols), index=data_encoded.index) # Use original index

# Concatenate the encoded features with the numerical features
# Drop the original categorical columns
numerical_data = data_encoded.drop(columns=categorical_cols)

# Reset index before concatenating to ensure unique index values
numerical_data = numerical_data.reset_index(drop=True)
encoded_df = encoded_df.reset_index(drop=True)

final_data = pd.concat([numerical_data, encoded_df], axis=1)

# Now, use final_data for IsolationForest
clf = IsolationForest(random_state=10, contamination=0.01)
clf.fit(final_data)
y_pred_outliers = clf.predict(final_data)

# Identify outlier indices
outlier_indices = data_encoded[y_pred_outliers == -1].index
```

```python
# Remove outliers from the dataset
data_no_outliers = data_encoded.drop(outlier_indices)
```
