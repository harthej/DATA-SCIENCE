```python
def recommend_anime(anime_id, user_sim_df, anime_df, threshold=0.8):
  """Recommends anime based on cosine similarity.


  # Get the similarity scores for the target anime.
  similarity_scores = user_sim_df.loc[anime_id]

  # Filter out anime with similarity scores below the threshold.
  similar_anime = similarity_scores[similarity_scores >= threshold]

  # Get the IDs of the similar anime.
  recommended_anime = similar_anime.index.tolist()

  # Remove the target anime from the recommendations.
  recommended_anime.remove(anime_id)

  # Return the list of recommended anime IDs.
  return recommended_anime
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, f1_score

# Split the dataset into training and testing sets.
train_data, test_data = train_test_split(anime_df, test_size=0.2, random_state=42)

# Create a user-anime matrix for the training set.
train_user_anime_df = train_data.pivot(index='anime_id',
                                      columns='name',
                                      values='rating').reset_index(drop=True)
# Get the unique anime IDs present in the training data
train_anime_ids = train_data['anime_id'].unique()
# Set the index using the unique anime IDs from the training data
train_user_anime_df.index = train_anime_ids
train_user_anime_df.fillna(0, inplace=True)

# Calculate cosine similarity for the training set.
train_user_sim = 1 - pairwise_distances(train_user_anime_df.values, metric='cosine')
train_user_sim_df = pd.DataFrame(train_user_sim)
# Set the index and columns using the unique anime IDs from the training data
train_user_sim_df.index = train_anime_ids
train_user_sim_df.columns = train_anime_ids
np.fill_diagonal(train_user_sim, 0)

# Make recommendations for the testing set.
recommendations = []
for anime_id in test_data['anime_id'].unique():
  # Check if anime_id exists in train_user_sim_df before making recommendations
  if anime_id in train_user_sim_df.index:
    recommendations.extend(recommend_anime(anime_id, train_user_sim_df, train_data, threshold=0.8))

# Evaluate the recommendations using precision, recall, and F1-score.
# Filter test_data to only include anime IDs present in train_user_sim_df
relevant_test_data = test_data[test_data['anime_id'].isin(train_user_sim_df.index)]
precision = precision_score(relevant_test_data['anime_id'].tolist(), recommendations, average='weighted')
recall = recall_score(relevant_test_data['anime_id'].tolist(), recommendations, average='weighted')
f1 = f1_score(relevant_test_data['anime_id'].tolist(), recommendations, average='weighted')

print(f"Precision: {precision}")
print(f"Recall: {recall}")
print(f"F1-score: {f1}")
```
