```python
import seaborn as sns
cols = data_cleaned1.columns
colours = ['#000099', '#ffff00'] # specify the colours - yellow is missing. blue is not missing.
sns.heatmap(data_cleaned1[cols].isnull(),
            cmap=sns.color_palette(colours))
```

```python
data_cleaned1[data_cleaned1.isnull().any(axis=1)].head()
```

```python
data_cleaned1.isnull().sum()
```

```python
#Mean Imputation
mean = data_cleaned1['AC'].mean()
print(mean)
```

```python
data_cleaned1['AC'] = data_cleaned1['AC'].fillna(mean)
```

```python
data_cleaned1
```

```python
#Missing value imputation for categorical vlaue
#Get the object columns
obj_columns=data_cleaned1[['AC']]
```

```python
obj_columns.isnull().sum()
```

```python
#Missing value imputation for categorical vlaue
obj_columns=obj_columns.fillna(obj_columns.mode().iloc[0])
```

```python
data_cleaned2=pd.concat([data_cleaned1,obj_columns],axis=1)
```

```python
plt.scatter(data_cleaned1.AC,data_cleaned1.FM)
```

```python
#Correlation
data_cleaned2.iloc[:,:-1].corr(method='pearson')
```
